#### Plot everything

```python
proposal = distributions.Gaussian(0, 115)
target = distributions.Gaussian(0, 1)
sizes = [4, 4, 4]
```

### Train

```python
learning_rate = 0.01
learner = kernel_learning.KernelLearner(key, target, sizes, kernels.get_rbf_kernel(1), learning_rate, lambda_reg=0, normalizing_constant=False)
```

```python
print("Training kernel to optimize KSD...")
learner.train(proposal, n_steps=100, batch_size=400, sample_every=True)
```

```python
learning_rate = 1
learner_high_lr = kernel_learning.KernelLearner(key, target, sizes, kernels.get_rbf_kernel(1), learning_rate, lambda_reg=0, normalizing_constant=False)

print("Training kernel to optimize KSD...")
learner_high_lr.train(proposal, n_steps=100, batch_size=400, sample_every=True)
```

```python
fig, axs = plt.subplots(2, 2, figsize=[18, 8])
for ax, l in zip(axs[0], [learner, learner_high_lr]):
    ax.plot(l.rundata["training_ksd"])

for ax, l in zip(axs[1], [learner, learner_high_lr]):
    ax.plot(l.rundata["bandwidth"])
```
